```python
import itertools
import math
import random
import pandas as pd
import os
import numpy
import time 
import matplotlib.pyplot as plt
import timeit


#rumus pyhtagoras untuk menghitung jarak antar dua kota
#menggunakan titik koordinat
def jarak(x1, y1, x2, y2) :
  return math.sqrt((x1-x2)**2+(y1-y2)**2)

#fungsi menghhitung jarak rute perjalanan
def totalJarak(tour) :
  d=0
  #Hitung jarak anatar kota dari kota-1 sampai kota-n 
  for i in range(1,len(tour)):
        x1 = wisata[tour[i-1]][0]
        y1 = wisata[tour[i-1]][1]
        x2 = wisata[tour[i]][0]
        y2 = wisata[tour[i]][1]
        d += jarak(x1,y1,x2,y2)
  #hitung jarak antara kota pertama dan kota terakhir
  x1 = wisata[tour[len(tour)-1]][0]
  y1 = wisata[tour[len(tour)-1]][1]
  x2 = wisata[tour[0]][0]
  y2 = wisata[tour[0]][1]
  d += jarak(x1,y1,x2,y2)
  return d

wisata = [[23,70],[33,29],
          [13,36],[65,21],
          [64,56],[53,93],
          [21,5],[63,93],
          [41,79],[59,94],
          [75,41],[75,52],
          [61,67],[47,60],
          [32,60]]
n = int(input("Masukan jumlah wisata (1 - 14): "))


x = []
y = []
x.append(0)
y.append(0)

t0 = time.perf_counter()

for k in range(1,n+1):
    x.append(k)

    #Generate SATU Rute perjalanan
    tour = random.sample(range(k),k)

    #Membuat SEMUA Rute Perjalanan yang memungkinkan
    generateTour = list(itertools.permutations(tour))

    jarakTerpendek = 999999
    ruteTerpendek = tour
    for i in range(len(generateTour)):
        j = totalJarak(generateTour[i])
        if j<jarakTerpendek:
            jarakTerpendek = int(j)
            ruteTerpendek = i
    y.append(time.perf_counter()-t0)

    print("Jarak terpendek",k,"destinasi wisata yaitu",jarakTerpendek,"KM")
    print("    Dengan Rute :",generateTour[i])

plt.plot([wisata[tour[i % n]][0] for i in range(n+1)],
[wisata[tour[i % n]][1] for i in range(n+1)], 'xb-' )
print()
fig, ax = plt.subplots()
ax.plot(x,y,'ro-')
ax.set(xlabel = 'Number of Destinasi',ylabel = 'Times (s)')
ax.set(xticks = list(range(0,10)))
ax.grid()
plt.show()

    #plt.plot([wisata[tour[i % n]][0] for i in range(n+1)],[wisata[tour[i % n]][1] for i in range(n+1)], 'xb-' )

```
